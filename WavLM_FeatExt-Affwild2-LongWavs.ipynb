```python
import soundfile as sf
import librosa

for file in file_list:

    data , samplerate = sf.read( file )
    print ("--------------")
    print ("Sample Rate: " + str(samplerate) + " Length: " + str(data.shape) + " " + "Time: " + str(data.shape[0]/samplerate) + " sec " + str( file ) )

    file_id = 'features/test/'+file.split('/')[3].split('.')[0]+'.wavlm' 
    
    if not os.path.exists(file_id):     
    
        # Calculate the duration of the audio file in seconds
        duration = librosa.get_duration(y = data, sr = samplerate)

        # Check if the duration is more than 400 seconds
        
        if len(data) > 6400000:
        #if duration > 6400000:
            # Calculate the number of segments needed
            num_segments = int(duration / 400) + 1
    
            # Split the audio into segments of 400 seconds or less
            for i in range(num_segments):
                start = i * 6400000
                end = min((i + 1) * 6400000, len(data))
                segment = data[start:end]
                print("Segment length: "+str(len(segment))+" "+str(file) )

                # extract the representation of last layer
                wav_input = torch.from_numpy(segment).float()
                wav_input_16khz = torch.unsqueeze(wav_input,0)

                del wav_input
                del segment
                torch.cuda.empty_cache()
        
                if cfg.normalize:
                    wav_input_16khz = torch.nn.functional.layer_norm(wav_input_16khz , wav_input_16khz.shape)
    
                rep = model.extract_features(wav_input_16khz)[0]

                del wav_input_16khz
                torch.cuda.empty_cache()
        
                rep[0].shape
                rep_np = rep[0].detach().numpy()

                del rep

                # Check if rep_df already exists
                if 'rep_df' in locals() or 'rep_df' in globals():
                    # If rep_df exists, append it with rep_np
                    rep_df = pd.concat([rep_df, pd.DataFrame(rep_np)], ignore_index=True)
                else:
                    # If rep_df does not exist, create it from rep_np
                    rep_df = pd.DataFrame(rep_np)
                    del rep_np

            print(f"The audio file has been split into {num_segments} segments of {duration/num_segments} seconds")

            file_id = 'features/test/'+file.split('/')[3].split('.')[0]+'.wavlm'  
            rep_df.to_csv(file_id)

            del rep_df
            del data

        else:
            print("The duration of the audio file is within the limit.")
```

```python
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
# create classifier and grouping object
clf = make_pipeline(
    StandardScaler(), 
    SVC(gamma='auto'),
)
logo = LeaveOneGroupOut()

def experiment(
    features,
    targets,
    groups,
):        
    truths = []
    preds = []
    
    # leave-one-speaker loop    
    pbar = audeer.progress_bar(
        total=len(groups.unique()),
        desc='Run experiment',
    )
    for train_index, test_index in logo.split(
        features, 
        targets, 
        groups=groups,
    ):
        train_x = features.iloc[train_index]
        train_y = targets[train_index]
        clf.fit(train_x, train_y)
        
        truth_x = features.iloc[test_index]
        truth_y = targets[test_index]
        predict_y = clf.predict(truth_x)
        
        truths.append(truth_y)
        preds.append(predict_y)
        
        pbar.update()
        
    # combine speaker folds
    truth = pd.concat(truths)
    truth.name = 'truth'
    pred = pd.Series(
        np.concatenate(preds),
        index=truth.index,
        name='prediction',
    )
    
    return truth, pred
```
